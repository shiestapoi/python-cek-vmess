### Case 4 - Connectivity check akurat (recommended)
**Kapan dipakai:** verifikasi apakah koneksi VMess benar-benar bisa dipakai lewat `xray`.

**Command:**
```bash
!PATH="$PWD/xray-bin:$PATH" python vmess_report.py --check-connectivity --connect-timeout 10 --connect-workers 40
```

**Interpretasi status connectivity:**
- `ok`: IP keluar cocok dengan endpoint valid.
- `not matched`: koneksi ada, tapi IP keluar tidak sesuai endpoint.
- `failed`: gagal koneksi/proxy dalam batas waktu.
- `skipped`: entry tidak layak diuji (status awal bukan `ok`).

````python
!PATH="$PWD/xray-bin:$PATH" python vmess_report.py --check-connectivity --connect-timeout 10 --connect-workers 40
````

## Troubleshooting Singkat
- Jika muncul `xray core not found in PATH`, ulangi cell install xray lalu jalankan command dengan prefix `PATH="$PWD/xray-bin:$PATH"`.
- Jika `report-only` gagal, pastikan file `report.json` memang sudah ada.
- Untuk environment non-interaktif, tambahkan `--no-progress` agar output lebih ringkas.
